```python
import pandas as pd
```

```python
df = pd.read_csv("./archive/dataset.csv")
df.head()
```

```python
target = df["Disease"]
df.drop("Disease",axis=1,inplace=True)
```

```python
from numpy import NaN

symptoms = []
for index, row in df.iterrows():
    symptomsRow = dict()
    for i, v in row.items():
        if(pd.notna(v)):
            symptomsRow[v] = 1
    symptoms.append(symptomsRow)

df2 = pd.DataFrame(symptoms)
df2.fillna(0,inplace=True)
df2["Disease"] = target
df2.to_csv("./data/symptoms.csv")
df2
```

Changing string Disease label to int

```python
from sklearn import preprocessing

le = preprocessing.LabelEncoder()
le.fit(df2.Disease)
df2['Disease'] = le.transform(df2.Disease)
df2
```

Storing Mapping Table

```python
diseaseDf = pd.DataFrame(columns = ['Disease', 'Value'])
diseaseDf["Disease"] = target.unique()
diseaseDf["Value"] = df2["Disease"].unique()
diseaseDf.to_csv("DiseaseMapping.csv")
```
